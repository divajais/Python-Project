Creating the new list having necessary data as per the requirements

```python
album_list=[]
for row in data['items']:
    album_id=row['track']['album']['id']
    album_name=row['track']['album']['name']
    release_date=row['track']['album']['release_date']
    total_tracks=row['track']['album']['total_tracks']
    artists_id=row['track']['album']['artists'][0]['id']
    artists_name=row['track']['album']['artists'][0]['name']
    disc_number=row['track']['disc_number']
    duration_ms=row['track']['duration_ms']
    track_name=row['track']['name']
    popularity=row['track']['popularity']
    track_number=row['track']['track_number']
    album_element={'album_id':album_id,'album_name':album_name,'release_date':release_date,'total_tracks':total_tracks,
                   'artists_id':artists_id,'artists_name':artists_name,'disc_number':disc_number,'duration_ms':duration_ms,
                  'track_name':track_name,'popularity':popularity,'track_number':track_number}
    album_list.append(album_element)
```

```python
album_list
```

```python
import pandas as pd
album_df=pd.DataFrame.from_dict(album_list)#Converting the data to DataFrame using Pandas
```
